<h4>En este proyecto, asumiremos el rol de un gerente comercial de la cadena SuperSales, una hipertienda con presencia en todo Estados Unidos. Nuestro objetivo es identificar los puntos clave para mejorar y elevar las ganancias de la empresa.
Como gerente, nuestra responsabilidad es maximizar las ventas y optimizar los beneficios en todas las regiones. Utilizaremos análisis de datos y herramientas de visualización para obtener información valiosa sobre las categorías de productos, las subcategorías más vendidas y las tendencias de ventas por región. Con estos insights, podremos desarrollar estrategias efectivas y tomar decisiones informadas para aumentar las ganancias de SuperSales.</h4>

## Importación del Dataset y paquetes a utilizar

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('data/SampleSuperstore.csv')
df
```

```
print('Dimensión del DataFrame')
print(df.shape)
```

```
print('Búsqueda de valores nulos')
print(df.isnull().sum())
```

```
print('Tipo de dato por columna')
print(df.dtypes)
```

## Análisis de ventas

<h3>Ventas por categoría</h3>

```
sns.barplot(x='Category', y='Sales', data=df, estimator=sum, errorbar=None)
plt.title('Ventas por Categoría')
plt.xlabel('Ventas')
plt.ylabel('Categoría')
plt.show()
```

<h3>Ganancias por Categoría</h3>

```
sns.barplot(x='Category', y='Profit', data=df, estimator=sum, errorbar=None)
plt.title('Ganancias por Categoría')
plt.xlabel('Ganancias')
plt.ylabel('Categoría')
plt.show()
```

<h4>En primer lugar, al examinar los ingresos totales, podemos observar que la categoría de Tecnología es la que genera la mayor cantidad de ingresos, seguida de cerca por Muebles y Materiales de Oficina. Sin embargo, la diferencia más significativa se encuentra en las ganancias, donde la categoría de Muebles genera considerablemente menos ganancias en comparación con las otras categorías.</h4>
<h4>Para comprender mejor por qué la categoría de Muebles genera tan pocas ganancias, procederemos a evaluar las ventas y las ganancias por subcategoría dentro de esta categoría en particular.</h4>

```
sns.barplot(x='Sales', y='Sub-Category', data=df, estimator=sum, errorbar=None, hue='Category')
plt.title('Ventas por Sub Categoría')
plt.xlabel('Ventas')
plt.ylabel('Categoría')
plt.show()
```

### Ganancias por SubCategorías

```
sns.barplot(x='Profit', y='Sub-Category', data=df, estimator=sum, errorbar=None, hue='Category')
plt.title('Ganancias por Sub Categoría')
plt.xlabel('Ganancias')
plt.ylabel('Categoría')
plt.show()
```

<h4>Al examinar las subcategorías de productos, se confirma que los artículos de tecnología, especialmente los teléfonos, mantienen un alto volumen de ventas, lo que los convierte en la categoría más rentable para nuestra empresa.</h4>
<h4>Además, observamos que los muebles también tienen una cantidad significativa de ventas, principalmente en el caso de las sillas. Sin embargo, nos enfrentamos a grandes pérdidas debido a las mesas, seguidas en menor medida por las bibliotecas.</h4>
<h4>Es de suma importancia para la empresa identificar la causa de estas pérdidas y tomar medidas correctivas de manera inmediata. Continuaremos explorando nuestros datos para evaluar este aspecto en particular, así como otras estrategias que nos ayuden a mejorar nuestras ganancias.</h4>

### Ventas y Ganancias por Región

```
ventas_por_region = df.groupby('Region')['Sales'].sum()
plt.pie(ventas_por_region, labels=ventas_por_region.index, autopct='%1.1f%%')
plt.title('Ventas por Región')
plt.axis('equal')
plt.show()
```

```
ganancia_por_region = df.groupby('Region')['Profit'].sum()
plt.pie(ganancia_por_region, labels=ganancia_por_region.index, autopct='%1.1f%%')
plt.title('Ganancia por Región')
plt.axis('equal')
plt.show()
```

<h4>Al analizar las ventas por región, podemos observar que la región Oeste es la que genera la mayor cantidad de ventas, seguida por la región Este. La región Central se ubica en tercer lugar, mientras que la región Sur ocupa el último puesto. Sin embargo, al evaluar las ganancias, notamos que a pesar de que la región Central representa el 21.8% de las ventas totales, solo contribuye con el 13.9% de las ganancias.</h4>
<h4>Con base en estos hallazgos, es crucial centrar nuestros esfuerzos en comprender qué factores hacen que la región Oeste sea la más rentable. A partir de ese conocimiento, podremos desarrollar estrategias para mejorar las ventas en la región Sur y aumentar las ganancias en la región Central.</h4>

```
# Filtrar por la región "West"
df_west = df[df['Region'] == 'West']

# Calcular las ventas por categoría
sales_by_category = df_west.groupby('Category')['Sales'].sum().reset_index()

# Crear el gráfico de barras
sns.barplot(y='Sales', x='Category', data=sales_by_category, color='steelblue')

# Establecer título y etiquetas de los ejes
plt.title('Ventas por categoría (Región Oeste)')
plt.xlabel('Ventas')
plt.ylabel('Categoría')

# Mostrar el gráfico
plt.show()
```

```
# Calcular las ganancias por categoría
profit_by_category = df_west.groupby('Category')['Profit'].sum().reset_index()

# Crear el gráfico de barras
sns.barplot(y='Profit', x='Category', data=profit_by_category, color='steelblue')

# Establecer título y etiquetas de los ejes
plt.title('Ganancia por categoría (Región Oeste)')
plt.xlabel('Ganancia')
plt.ylabel('Categoría')

# Mostrar el gráfico
plt.show()
```

```
# Contar las ventas por categoría
category_counts_w = df_west['Category'].value_counts()

# Crear el gráfico de barras
sns.barplot(x=category_counts_w.index, y=category_counts_w.values)
plt.title('Categorías más vendidas en la región Oeste')
plt.xlabel('Categoría')
plt.ylabel('Ventas')
plt.xticks(rotation=45)
plt.show()
```

```
# Agrupar por subcategoría y sumar las ganancias
subcat_units_w = df_west.groupby('Sub-Category')['Quantity'].sum().reset_index()

# Crear el gráfico de barras
sns.barplot(x='Quantity', y='Sub-Category', data=subcat_units_w)
plt.title('Unidades vendidas por subcategoría en la región Oeste')
plt.xlabel('Unidades vendidas')
plt.ylabel('Subcategoría')
plt.show()
```

```
# Agrupar por subcategoría y sumar las unidades vendidas
subcat_profit_w = df_west.groupby('Sub-Category')['Profit'].sum().reset_index()

# Crear el gráfico de barras
sns.barplot(x='Profit', y='Sub-Category', data=subcat_profit_w)
plt.title('Ganancia por subcategoría en la región Oeste')
plt.xlabel('Ganancia')
plt.ylabel('Subcategoría')
plt.show()
```

<h4><h4>Al analizar las ventas en la región Oeste, podemos destacar que la categoría de Artículos de Oficina es la que genera mayores ventas y ganancias, principalmente impulsada por las ventas de carpetas y papel.</h4>
<h4>Además, es importante destacar que las Fotocopiadoras representan una fuente significativa de ganancias, a pesar de no venderse en grandes cantidades.</h4>
<h4>Ahora realizaremos un análisis similar en las demás regiones con el objetivo de identificar las diferencias que nos ayuden a desarrollar estrategias para mejorar las ventas y ganancias en nuestras regiones menos rentables.</h4>

```
# Filtrar por la región "South"
df_south = df[df['Region'] == 'South']

# Calcular las ventas por categoría
sales_by_category_s = df_south.groupby('Category')['Sales'].sum().reset_index()

# Crear el gráfico de barras
sns.barplot(y='Sales', x='Category', data=sales_by_category_s, color='steelblue')

# Establecer título y etiquetas de los ejes
plt.title('Ventas por categoría (Región Sur)')
plt.xlabel('Ventas')
plt.ylabel('Categoría')

# Mostrar el gráfico
plt.show()
```

```
# Calcular las ganancias por categoría
profit_by_category_s = df_south.groupby('Category')['Profit'].sum().reset_index()

# Crear el gráfico de barras
sns.barplot(y='Profit', x='Category', data=profit_by_category_s, color='steelblue')

# Establecer título y etiquetas de los ejes
plt.title('Ganancia por categoría (Región Sur)')
plt.xlabel('Ganancia')
plt.ylabel('Categoría')

# Mostrar el gráfico
plt.show()
```

```
# Contar las ventas por categoría
category_counts_s = df_south['Category'].value_counts()

# Crear el gráfico de barras
sns.barplot(x=category_counts_s.index, y=category_counts_s.values)
plt.title('Categorías más vendidas en la región Sur')
plt.xlabel('Categoría')
plt.ylabel('Ventas')
plt.xticks(rotation=45)
plt.show()
```

```
# Agrupar por subcategoría y sumar las ganancias
subcat_units_s = df_south.groupby('Sub-Category')['Quantity'].sum().reset_index()

# Crear el gráfico de barras
sns.barplot(x='Quantity', y='Sub-Category', data=subcat_units_s)
plt.title('Unidades vendidas por subcategoría en la región Sur')
plt.xlabel('Unidades vendidas')
plt.ylabel('Subcategoría')
plt.show()
```

```
# Agrupar por subcategoría y sumar las unidades vendidas
subcat_profit_s = df_south.groupby('Sub-Category')['Profit'].sum().reset_index()

# Crear el gráfico de barras
sns.barplot(x='Profit', y='Sub-Category', data=subcat_profit_s)
plt.title('Ganancia por subcategoría en la región Sur')
plt.xlabel('Ganancia')
plt.ylabel('Subcategoría')
plt.show()
```

```
# Filtrar el DataFrame por las regiones oeste y sur
df_west = df[df['Region'] == 'West']
df_south = df[df['Region'] == 'South']

# Contar el número de clientes por región
customer_count = pd.DataFrame({
    'Region': ['West', 'South'],
    'CustomerCount': [df_west.shape[0], df_south.shape[0]]
})

# Crear el gráfico de barras
sns.barplot(x='Region', y='CustomerCount', data=customer_count, color='steelblue')

# Establecer título y etiquetas de los ejes
plt.title('Número de clientes (Región Oeste vs. Región Sur)')
plt.xlabel('Región')
plt.ylabel('Número de clientes')

# Mostrar el gráfico
plt.show()
```


<h4>Al comparar la región Sur con la región Oeste, se puede concluir que las ventas promedio de muebles son más bajas en la región Sur. Además, se observa que las Mesas en la región Sur generan pérdidas significativas. Estos factores seguramente influyen en el porcentaje de ganancias de la empresa, por lo que es importante abordar esta situación de manera inmediata.</h4>
<h4>Por otro lado, se destaca que la región Oeste cuenta con aproximadamente el doble de clientes en comparación con la región Sur. Por lo tanto, se sugiere implementar campañas publicitarias y otras estrategias para atraer a más clientes y mejorar los resultados en esta región.</h4>

```
# Contar las ventas por categoría
category_counts_c = df_central['Category'].value_counts()

# Crear el gráfico de barras
sns.barplot(x=category_counts_c.index, y=category_counts_c.values)
plt.title('Categorías más vendidas en la región Central')
plt.xlabel('Categoría')
plt.ylabel('Ventas')
plt.xticks(rotation=45)
plt.show()
```

```
# Filtrar el DataFrame por las regiones oeste y central
df_west = df[df['Region'] == 'West']
df_central = df[df['Region'] == 'Central']

# Contar el número de clientes por región
customer_count = pd.DataFrame({
    'Region': ['West', 'Central'],
    'CustomerCount': [df_west.shape[0], df_central.shape[0]]
})

# Crear el gráfico de barras
sns.barplot(x='Region', y='CustomerCount', data=customer_count, color='steelblue')

# Establecer título y etiquetas de los ejes
plt.title('Número de clientes (Región Oeste vs. Región Central)')
plt.xlabel('Región')
plt.ylabel('Número de clientes')

# Mostrar el gráfico
plt.show()
```

<h4>Al analizar esta región, se observa que, aunque se registra una menor cantidad de ventas en comparación con la región Sur, la diferencia no es tan significativa. Sin embargo, se destaca que en esta región en particular, no se obtiene una ganancia considerable en la categoría de Artículos de Oficina, a pesar de ser una de las fortalezas de la Región Oeste. Además, se evidencian pérdidas significativas en el área de Muebles.</h4>
<h4>Ahora nos enfocaremos en determinar la causa de estas circunstancias utilizando los datos disponibles.</h4>


<h3>Clientes por región</h3>

```
# Agrupar por región y contar la cantidad de registros (clientes)
region_counts = df.groupby('Region').size()

# Crear el gráfico de barras
sns.barplot(x=region_counts.index, y=region_counts.values)
plt.title('Clientes por región')
plt.xlabel('Región')
plt.ylabel('Cantidad de clientes')
plt.show()
```


<h4>Continuando con nuestro análisis, hemos observado que la cantidad de clientes tiene un impacto directo en las ventas, lo cual resalta la importancia de mejorar este aspecto.</h4>
<h4>Ahora vamos a evaluar si el segmento de cliente también juega un papel influyente en nuestras ventas.</h4>

```
# Agrupar por región y tipo de segmento, y contar las ocurrencias
region_segment_counts = df.groupby(['Region', 'Segment']).size().unstack()

# Calcular el porcentaje de cada segmento por región
region_segment_percentages = region_segment_counts.apply(lambda x: x / x.sum() * 100, axis=1)

# Crear el gráfico de barras apiladas
region_segment_percentages.plot(kind='bar', stacked=True)
plt.title('Tipos de segmento por región')
plt.xlabel('Región')
plt.ylabel('Porcentaje')
plt.legend(title='Segmento')
plt.show()
```

<h4>Según nuestros análisis, parece que los segmentos de clientes no tienen un impacto significativo en las ventas ni en las ganancias.</h4>
<h4>Ahora procederemos a evaluar el efecto de los descuentos para determinar si tienen alguna influencia en nuestros resultados.</h4>

```
# Agrupar por región y calcular el promedio de descuentos
region_discounts = df.groupby('Region')['Discount'].mean().reset_index()

# Crear el gráfico de barras
sns.barplot(x='Region', y='Discount', data=region_discounts)
plt.title('Descuentos promedio por región')
plt.xlabel('Región')
plt.ylabel('Descuento promedio')
plt.show()
```

<h4>En este caso, hemos identificado el principal factor que influye en las bajas ganancias de nuestro sector central: la alta frecuencia de descuentos promedio aplicados en esta área.</h4>

```
# Filtrar el DataFrame por la subcategoría "mesas"
df_mesas = df[df['Sub-Category'] == 'Tables']

# Crear el gráfico de dispersión
sns.scatterplot(x='Discount', y='Profit', data=df_mesas)

# Establecer título y etiquetas de los ejes
plt.title('Relación entre Descuentos y Ventas de Mesas')
plt.xlabel('Descuento')
plt.ylabel('Ventas')

# Mostrar el gráfico
plt.show()
```

<h4>Al realizar una evaluación detallada de las ventas de Mesas, hemos observado que este producto tiene un margen de ganancia muy bajo. Incluso al aplicar cualquier tipo de descuento, nos genera pérdidas significativas.</h4>

## Conclusiones

<h4>Después de nuestra investigación, hemos llegado a las siguientes conclusiones:</h4>
<h4>Para mejorar nuestras ventas, es crucial enfocar nuestros recursos en aumentar el número de clientes, especialmente en las zonas Sur y Central.</h4>
<h4>Por otro lado, enfrentamos un gran desafío en términos de pérdidas en el área de Muebles, principalmente debido a las Mesas y Bibliotecas, que presentan un margen de ganancia reducido y se ven afectadas por descuentos significativos, especialmente en la región Central.</h4>
<h4>Por lo tanto, recomendamos revisar los costos de producción de estos productos para asegurar un margen rentable, y considerar la suspensión de descuentos en esta categoría. En cambio, podríamos redirigir los descuentos hacia la categoría de Tecnología, ya que productos como las Fotocopiadoras tienen un bajo volumen de ventas, pero generan excelentes ganancias.</h4>
<h4>Además, es necesario realizar un análisis detallado de cada región para identificar las tiendas específicas que presentan las ventas más bajas y los descuentos más altos.</h4>
